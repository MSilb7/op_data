```python
import pandas as pd
import plotly.express as px
import numpy as np
import os


df = pd.read_csv('csv_exports/questers_data/op_questers_stats_22_01_18.csv')
df = df.reset_index()
print(df.columns)
# display(df)
df.head()
```

```python
#De-Sybil
df_human = df[df['address_label'] == 'Potential Human']
print(len(df_human))
#News Only
df_noobs = df_human[df_human['address_age_label'] == 'New Address']
print(len(df_noobs))
```

```python


tx_pivot_table = df_noobs.pivot_table(
        values='num_txs_since_quests_off', index='recurring_label',
        columns=['churn_label'], aggfunc={'num_txs_since_quests_off':'sum'}
        ,margins = True)
addr_pivot_table = df_noobs.pivot_table(
        values='num_txs_since_quests_off', index='recurring_label',
        columns=['churn_label'], aggfunc={'num_txs_since_quests_off':'count'}
        ,margins = True)
# pivot_table.reset_index()
# pivot_table['percent_of_total'] = pivot_table.apply(lambda x: x/x.sum()*100, axis = 1)
# pivot_table['percent_of_txs'] = pivot_table['num_txs_since_quests_off'] / pivot_table['num_txs_since_quests_off'].sum() * 100
```
